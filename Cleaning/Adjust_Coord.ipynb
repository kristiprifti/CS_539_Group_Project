```python
df['coord'] = df['coord'].apply(ast.literal_eval)
```

```python
df['coord_rounded'] = df['coord'].apply(lambda x: (round(x[0], 2), round(x[1], 2)))
coord_list_rounded = [(round(x[0], 2), round(x[1], 2)) for x in coord_list]
```

```python
df[['latitude', 'longitude']] = pd.DataFrame(df['coord'].tolist(), index=df.index)

```

```python
df['latitude_rounded'] = df['latitude'].round(2)
df['longitude_rounded'] = df['longitude'].round(2)
```

```python
for i, row in df.iterrows():
    coord_rounded = (row['latitude_rounded'], row['longitude_rounded'])
    if coord_rounded in coord_list_rounded:
        new_coord = coord_list[coord_list_rounded.index(coord_rounded)]
        df.loc[i, 'latitude'] = new_coord[0]
        df.loc[i, 'longitude'] = new_coord[1]
```

```python
df = df.drop(['coord', 'coord_rounded', 'latitude_rounded', 'longitude_rounded'], axis=1)

df.head()
```
